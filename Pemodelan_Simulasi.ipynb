# Import Library dan inisiasi list

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from prettytable import PrettyTable
import pandas as pd
import bokeh
from bokeh.io import output_file, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, Panel, Tabs
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel
from bokeh.models.tools import HoverTool

 
Analitik_X = []
Analitik_Y = []
Analitik_hambatan_X = []
Analitik_hambatan_Y = []
Numerik_X = []
Numerik_Y = []
nilai_waktu_Analitik = []
nilai_waktu_Analitik_hambatan = []
nilai_waktu_Numerik = []
```

```python
def updateAnalitik(kec, sudut, deltaT):
  g = 9.8
  Sin = math.sin(math.radians(sudut))
  Cos = math.cos(math.radians(sudut))
  tTotal = float((2 * kec * Sin) / g)
  R = kec * Cos * tTotal
  posisi_X = 0
  posisi_Y = 0
  kec_X = float(kec * Cos)
  kec_Y = float(kec * Sin)
  g = g * -1
  a = float(math.pow(kec,2) * math.pow(Sin,2))
  b = float(-2*g)
  hMax = float(a/b)
  for t in np.arange(0,tTotal+deltaT,deltaT):
    posisi_X = kec_X * t
    Analitik_X.append(posisi_X)
    posisi_Y = kec_Y * t + (0.5 * g * math.pow(t,2))
    Analitik_Y.append(posisi_Y)
    nilai_waktu_Analitik.append(t)
    if posisi_Y < 0:
      print('Waktu dari solusi Analitik ketika menyentuh tanah: ',t)
      print("Tinggi Maksimum dari solusi Analitik: ", hMax)
      print('Jarak Maksimum dari solusi Analitik: ', R)
      print('\n')
      break
```

```python
def updateAnalitikHambatanUdara(kec, sudut, deltaT, D, m):
  g = 9.8
  Sin = math.sin(math.radians(sudut))
  Cos = math.cos(math.radians(sudut))
  tTotal = float((2 * kec * Sin) / g)
  R = kec * Cos * tTotal
  posisiX_hambatan = 0
  posisiY_hambatan = 0
  kec_X = float(kec * Cos)
  kec_Y = float(kec * Sin)
  g = g*-1
  for hambatan in np.arange(0,tTotal+deltaT,deltaT):
    z = math.pow(kec_X,2)
    y = math.pow(kec_Y,2)
    w = z+y
    v = math.sqrt(w)
    ax = -(D/m)*v*kec_X
    ay = g-(D/m)*v*kec_Y
    posisiX_hambatan = kec_X * hambatan + (0.5 * ax * math.pow(hambatan,2))
    Analitik_hambatan_X.append(posisiX_hambatan)
    posisiY_hambatan = kec_Y * hambatan + (0.5 * ay * math.pow(hambatan,2))
    Analitik_hambatan_Y.append(posisiY_hambatan)
    nilai_waktu_Analitik_hambatan.append(hambatan)
    if posisiY_hambatan < 0:
      print('Waktu dari solusi Analitik yang dipengaruhi hambatan udara ketika menyentuh tanah: ',hambatan)
      print("Tinggi Maksimum dari solusi Analitik yang dipengaruhi hambatan udara: ", max(Analitik_hambatan_Y))
      print('Jarak Maksimum dari solusi Analitik yang dipengaruhi hambatan udara: ', posisiX_hambatan)
      print('\n')
      break
```

```python
def updateNumerik(kec, sudut, deltaT):
  g = 9.8
  Sin = math.sin(math.radians(sudut))
  Cos = math.cos(math.radians(sudut))
  tTotal = float((2 * kec * Sin) / g)
  R = kec * Cos * tTotal
  posisi_X = 0
  posisi_Y = 0
  kec_X = float(kec * Cos)
  kec_Y = float(kec * Sin)
  g = g * -1
  for to in np.arange(0,tTotal+deltaT,deltaT):
      kec_Y = kec_Y + (g * deltaT)
      posisi_X += kec_X * deltaT
      posisi_Y += kec_Y * deltaT
      Numerik_X.append(posisi_X)
      Numerik_Y.append(posisi_Y)
      nilai_waktu_Numerik.append(to)
      if posisi_Y < 0:
        print('Waktu dari solusi Numerik ketika menyentuh tanah : ',to)
        print("Tinggi Maksimum dari solusi Numerik: ", max(Numerik_Y))
        print('Jarak Maksimal dari solusi Numerik ',posisi_X)
        print('\n')
        break
```

```python
print("===========================================================================")
print("                      TUBES 1 PEMODELAN DAN SIMULASI")
print("                               GERAK PELURU")
print("===========================================================================")
print()
print()
print("Karena ini adalah simulasi gerak peluru maka")
kec = float(input('Kecepatan: '))
sudut = float(input('Sudut Kemiringan: '))
deltaT = float(input('Delta T: '))
D = float(input('Konstanta Benda / D: '))
m = float(input('Massa Benda: '))
print()
def menu():
  print("--------------------------------   MENU   --------------------------------")
  print("1. Solusi Analitik")
  print("2. Solusi Analitik yang dipengaruhi hambatan udara")
  print("3. Solusi Numerik")
  print("4. Grafik")
  print("5. Tabel")
  print("6. Keluar")
menu()
pil = float(input("pilihan: "))
print()
 
while pil < 7:
  if pil == 1:
    updateAnalitik(kec, sudut, deltaT)
    menu()
    pil = float(input("pilihan: "))
    print()
  elif pil == 2:
    updateAnalitikHambatanUdara(kec, sudut, deltaT, D, m)
    menu()
    pil = float(input("pilihan: "))
    print()
  elif pil == 3:
    updateNumerik(kec, sudut, deltaT)
    menu()
    pil = float(input("pilihan: "))
    print()
  elif pil == 4:
    output_notebook()
    fig = figure(title='Grafik Perbandingan',
                         plot_height=400, plot_width=850)
    fig.line(x=Analitik_X, y=Analitik_Y,color='deepskyblue', line_width = 3, legend_label='Solusi Analitik')
    fig.line(x=Analitik_hambatan_X, y=Analitik_hambatan_Y, color='palegreen', line_width = 2, legend_label='Solusi Analitik dipengaruhi hambatan udara')
    fig.line(x=Numerik_X, y=Numerik_Y, color='coral', line_width = 2, legend_label='Solusi Numerik')
    fig.legend.location = 'top_left'
    show(fig)
    menu()
    pil = float(input("pilihan: "))
    print()
  elif pil == 5:
    tabel = {"Waktu" : nilai_waktu_Analitik, "JarakAnalitik": Analitik_X, "TinggiAnalitik" : Analitik_Y, "JarakNumerik": Numerik_X,
             "TinggiNumerik": Numerik_Y}
    tabelHambatan = {"Waktu": nilai_waktu_Analitik_hambatan,
             "Jarak": Analitik_hambatan_X, "Tinggi": Analitik_hambatan_Y}
    df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in tabel.items() ]))
    dfhambatan = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in tabelHambatan.items() ]))
    print("Tabel Analitik VS Numerik")
    print("-------------------------")
    print(df)
    print()
    print("Tabel solusi Analitik yang dipengaruhi oleh hambatan udara")
    print("----------------------------------------------------------")
    print(dfhambatan)
    menu()
    pil = float(input("pilihan: "))
    print()
  elif pil == 6:
    break
```
